# doctest

Sometimes useful as a way to both document how to use the code and test for the right answer.

```python
"""
The procedure print_an_a() prints an 'a'.

>>> print_an_a()
a

With print_character(c) you print any character.

>>> print_character('a')
a

"""

import doctest

def print_an_a():
    print_character('a')
    
def print_character(c: str):
    print(f'{c}')

    
if __name__ == '__main__':
    import doctest
    doctest.testmod()
```

# unittest

For any unit of code. Below is the simplest but tests can be bundled in test suites.

```python
import unittest


def multiply(x, y):
    return x * y


class MultiplyTest(unittest.TestCase):

    def test_3_times_4(self):
        self.assertEqual(multiply(3, 4), 12)
        
    def test_3_times_5(self):
        self.assertEqual(multiply(3, 5), 15)
        
    def another_test(self):
        self.assertEqual(1, 2)


# the arguments are not needed when invoked from the command line
if __name__ == '__main__':
    unittest.main(argv=[''], exit=False)
```
